```python
from sklearn.linear_model import LogisticRegression
import numpy as np
import os,sys
import cv2

train_path = '/content/eyegaze_new/train/'
val_path = '/content/eyegaze_new/val/'
X = np.zeros(shape =  (0,224*224*3))
y = np.zeros(shape = (0,1))

for folder in ['left', 'right']:
  images = os.listdir(os.path.join(train_path,folder))
  for i, imagename in enumerate(images):
    image = cv2.imread(os.path.join(train_path, folder, imagename))
    resized_image = cv2.resize(image, (224, 224), interpolation=cv2.INTER_CUBIC)
    resized_image = resized_image.reshape(1,-1)
    X = np.concatenate((X,resized_image),axis=0)
    if folder == 'left':
      value = 0
    else:
      value = 1
    y = np.concatenate((y,np.array([value]).reshape(1,1)),axis=0)
y = y.reshape(-1)
    
X_val = np.zeros(shape =  (0,224*224*3))
y_val = np.zeros(shape = (0,1))
for folder in ['left', 'right']:
  images = os.listdir(os.path.join(val_path,folder))
  for i, imagename in enumerate(images):
    image = cv2.imread(os.path.join(val_path, folder, imagename))
    resized_image = cv2.resize(image, (224, 224), interpolation=cv2.INTER_CUBIC)
    resized_image = resized_image.reshape(1,-1)
    X_val = np.concatenate((X_val,resized_image),axis=0)
    if folder == 'left':
      value = 0
    else:
      value = 1
    y_val = np.concatenate((y_val,np.array([value]).reshape(1,1)),axis=0)
y_val = y_val.reshape(-1)
print(X.shape)
print(X_val.shape, y_val.shape)

clf = LogisticRegression(random_state=0, solver='lbfgs', multi_class='multinomial').fit(X, y)
train_acc = clf.score(X,y)
clf.predict(X_val)

val_acc = clf.score(X_val, y_val)
print("train accuracy:{}".format(train_acc))
print("validation accuracy:{}".format(val_acc))
```

```python
from sklearn.svm import SVC
svm = SVC(gamma='auto')
svm.fit(X, y)
train_acc = svm.score(X,y)
val_acc = svm.score(X_val, y_val)
```

```python
print("train accuracy:{}".format(train_acc))
print("validation accuracy:{}".format(val_acc))
```
