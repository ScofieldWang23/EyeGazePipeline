```python
import os
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile
import shutil
```

```python
base_dir = os.getcwd()

# result photo directory
result_dir = os.path.join(base_dir, 'preparation')

# num of cameras
num_of_cameras = 2
```

```python
df = pd.read_csv(os.path.join(base_dir, 'gaze_data.csv'))
df.shape
```

```python
def move_to_result_directory(filename, source_directory, result_directory, sub_directory):
    print("move has been made!")
    # create directory if not exists
    if not os.path.exists(result_directory):
        os.makedirs(result_directory)
    
    if not os.path.exists(os.path.join(result_directory, sub_directory)):
        os.makedirs(os.path.join(result_directory, sub_directory))
    
    destination_file_name = os.path.join(result_directory, sub_directory, filename)
    source_filename = os.path.join(source_directory, filename)
#     os.rename(destination_file_name)
    shutil.copy2(source_filename, destination_file_name)
```

```python
def process_raw_directory(idx_of_camera, df, result_dir):
    raw_photo_directory_i_cam = os.path.join(base_dir, 'face' + str(idx_of_camera))
    # iterate photo directory and excel directory
    list_raw_photo_directory = os.listdir(raw_photo_directory_i_cam)
    list_raw_photo_directory.sort()
    delimeter = "_"
    delay_threshold = 0.01

    i = 0
    j = 0 
    while i < len(list_raw_photo_directory) and j < df.shape[0]:
        raw_photo_name = list_raw_photo_directory[i]
        gaze_record = df.iloc[j]
        
        gaze_point_timestamp = gaze_record['timestamp']
        raw_photo_timestamp = float(raw_photo_name.split(delimeter)[-1][:-4])
        if gaze_point_timestamp != gaze_point_timestamp:
            j += 1
        diff_time = gaze_point_timestamp - raw_photo_timestamp

        # if valid, with delay smaller than threshold
        if abs(diff_time) < delay_threshold:
            print(diff_time)

            # decide if left or right
            if gaze_record['FPOGX'] < 0.5:
                move_to_result_directory(raw_photo_name, raw_photo_directory_i_cam, result_dir, r'left')
            else:
                move_to_result_directory(raw_photo_name, raw_photo_directory_i_cam, result_dir, r'right')
                
#             print(idx_of_camera, gaze_point_timestamp, raw_photo_timestamp)     

        if gaze_point_timestamp < raw_photo_timestamp:
            j += 1
        else:
            i += 1
    

```

```python
for i in range(num_of_cameras):
    process_raw_directory(i, df, result_dir)
```
